```python
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import display, HTML

rs = np.random.RandomState(10)
```

## Plotting a formula

```python
x = [i * 0.01 for i in range(-500, 500)]
y = [x**3 for x in x]
sns.lineplot(x, y)
```

## Histogram

```python
dt = rs.uniform(0, 10, 100)
display(dt)

sns.distplot(dt)
```

```python
x = rs.uniform(0, 10, 100)
y = rs.uniform(0, 10, 100)
sns.jointplot(x, y)
```

## Plot average values

```python
data = {'x':[], 'y':[]}

# First data set.
for i in range(-10, 10):
    data['x'].append(i * 0.1)
    data['y'].append(i**2 * (-1 - rs.rand()))

# Second data set.
for i in range(-10, 10):
    data['x'].append(i * 0.1)
    data['y'].append(i**2 * (1 + rs.rand()))
    
d = pd.DataFrame(data)
display(d)

sns.relplot(x='x', y='y', data=d, kind="line")
```
